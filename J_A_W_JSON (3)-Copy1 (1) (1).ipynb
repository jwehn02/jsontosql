# Here we have to handle individual columns for MySQL to read:

+ If you feel neccessary you can adjust this and create a function based on each datatype by column and do a task. This would need to be evaluated because, your columns are case-by-case and if these data change then your function may fail without proper error handling. 
+ There is another issue to think about for example: if you lump columns based on datatype, you have a potential problem because one column may be float-string that needs converting abd if you read the label it should be downcast to integer instead. 
    + Memory management should be considered, while bulk loading and sending data
    + Date-Time information can be manipulated easily if you need the times, I will comment below for this
    + Sending the raw dataframe you need to be careful due to round trips to the DB
+ Some columns seem ok at first for processessing until you scan them in and see weird behavior at the row level and then an error occurs. 

+ `NaN` values are an issue with **`DATE`** and you need to do something with this
    + Either create a dummy date format `YYYY-MM-DD` that doesn't exists for these data 
    OR
        + You should replace empty `Nan` values with a `CreatedDate` column for consistency 
        
+ `BOOLEAN`: when these occured in a column and I saw a huge number of empty rows, I made `NaN` FALSE unless noted
    + Some cases I had to make a judgement call and replace with TEXT for schema do to column formatting 

```python
# Remove Useless Columns:                                      
useless_columns=['pointsAwardedDate','rewardsReceiptItemList','finishedDate','purchaseDate']

# DROP THESE COLUMNS:
new_subset_play= play.drop(useless_columns,axis=1)

new_subset_play.head() # was 52 columns

# Changed the NaN to 0
new_subset_play['discountedItemPrice']=new_subset_play['discountedItemPrice'].fillna(0)

# Converting from string int to -> float
new_subset_play['discountedItemPrice']=pd.to_numeric(new_subset_play['discountedItemPrice'])

# confirming int 
type(new_subset_play['discountedItemPrice'][2])


# we have what are some identifying item_numbers and NaNs
# set(new_subset_play['itemNumber'])

#Same process as above:
new_subset_play['itemNumber']=new_subset_play['itemNumber'].fillna(0)

# Change this to INT with downcast='signed'
new_subset_play['itemNumber']=pd.to_numeric(new_subset_play['itemNumber'],downcast='signed')

#making this float with 0.0 with .nafill()
new_subset_play['originalMetaBriteQuantityPurchased']=new_subset_play['originalMetaBriteQuantityPurchased'].fillna(0.0)


new_subset_play['originalMetaBriteQuantityPurchased']=pd.to_numeric(new_subset_play['originalMetaBriteQuantityPurchased']
                                                                    ,downcast='signed')

new_subset_play['rewardsReceiptItemList_pointsEarned']=new_subset_play['rewardsReceiptItemList_pointsEarned'].fillna(0.0)

new_subset_play['rewardsReceiptItemList_pointsEarned']=pd.to_numeric(new_subset_play['rewardsReceiptItemList_pointsEarned'])


new_subset_play['targetPrice']=new_subset_play['targetPrice'].fillna(0)

new_subset_play['targetPrice']=pd.to_numeric(new_subset_play['targetPrice'])


# This is 91% Null, what do I fill the value with since it is BOOL? I can do empty str or FALSE or Zero

# set(new_subset_play['competitiveProduct'])

# Checking to see proportion of data NaN to decide how to handle these
new_subset_play['competitiveProduct'].isnull().sum()/len(new_subset_play['competitiveProduct'])

new_subset_play['competitiveProduct']=new_subset_play['competitiveProduct'].fillna("Unknown")


# set(new_subset_play['competitiveProduct'])

new_subset_play['originalFinalPrice']=new_subset_play['originalFinalPrice'].fillna(0.0)

new_subset_play['originalFinalPrice']=pd.to_numeric(new_subset_play['originalFinalPrice'])


new_subset_play['originalMetaBriteItemPrice']=new_subset_play['originalMetaBriteItemPrice'].fillna(0.0)

new_subset_play['originalMetaBriteItemPrice']=pd.to_numeric(new_subset_play['originalMetaBriteItemPrice'])

# 99% Nan, I will do unknown string but can convert to BOOL if you want and change schema
set(new_subset_play['deleted'])
new_subset_play['deleted'].isnull().sum()/len(new_subset_play['deleted'])


# new_subset_play['deleted']=new_subset_play['deleted'].fillna(0)

set(new_subset_play['deleted'])
#Bool values here, I replaced NaN with False as a judgement call and easier to import
new_subset_play["deleted"].fillna(False,inplace=True)


new_subset_play['priceAfterCoupon']=new_subset_play['priceAfterCoupon'].fillna(0.0)

new_subset_play['priceAfterCoupon']=pd.to_numeric(new_subset_play['priceAfterCoupon'])


set(new_subset_play['preventTargetGapPoints'])
new_subset_play['preventTargetGapPoints'].fillna(False,inplace=True)

new_subset_play['userFlaggedNewItem'].fillna(False,inplace=True)

# set(new_subset_play['userFlaggedPrice' ])

new_subset_play['userFlaggedPrice']=new_subset_play['userFlaggedPrice'].fillna(0.0)

new_subset_play['userFlaggedPrice']=pd.to_numeric(new_subset_play['userFlaggedPrice'])

# 'userFlaggedPrice' 

new_subset_play['userFlaggedQuantity']
new_subset_play['userFlaggedQuantity']=new_subset_play['userFlaggedQuantity'].fillna(0)

new_subset_play['userFlaggedQuantity']=pd.to_numeric(new_subset_play['userFlaggedQuantity'],downcast='signed')
# new_subset_play['userFlaggedQuantity']


new_subset_play['finalPrice']=new_subset_play['finalPrice'].fillna(0.0)
new_subset_play['finalPrice']=pd.to_numeric(new_subset_play['finalPrice'])
# new_subset_play['finalPrice']


new_subset_play['itemPrice']=new_subset_play['itemPrice'].fillna(0.0)
new_subset_play['itemPrice']=pd.to_numeric(new_subset_play['itemPrice'])

# set(new_subset_play['quantityPurchased'])
new_subset_play['quantityPurchased']=new_subset_play['quantityPurchased'].fillna(0)
new_subset_play['quantityPurchased']=pd.to_numeric(new_subset_play['quantityPurchased'],downcast='signed')

# set(new_subset_play['bonusPointsEarned'])
new_subset_play['bonusPointsEarned']=new_subset_play['bonusPointsEarned'].fillna(0.0)
new_subset_play['bonusPointsEarned']=pd.to_numeric(new_subset_play['bonusPointsEarned'])


# set(new_subset_play['purchasedItemCount'])
new_subset_play['purchasedItemCount']=new_subset_play['purchasedItemCount'].fillna(0)
new_subset_play['purchasedItemCount']=pd.to_numeric(new_subset_play['purchasedItemCount'],downcast='signed')

new_subset_play['totalSpent']=new_subset_play['totalSpent'].fillna(0.0)
new_subset_play['totalSpent']=pd.to_numeric(new_subset_play['totalSpent'])

new_subset_play['partnerItemId']=new_subset_play['partnerItemId'].fillna(0)
# set(new_subset_play['partnerItemId'])
new_subset_play['partnerItemId']=pd.to_numeric(new_subset_play['partnerItemId'])
```
